```python
#import library untuk analisis dataa

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv('Stocks.csv')
df
```

```python
# Distribusi df pada setiap atribut Purchase Price, Current Price, dan Gain/Loss dengan grafik histogram
plt.figure(figsize=(10, 4))

plt.subplot(1, 3, 1)
sns.histplot(df['Purchase Price ($)'], kde=True, bins=10, edgecolor='k', color="blue")
plt.xlabel('Purchase Price')
plt.ylabel('Frequency')
plt.title('Histogram of Purchase Price')

plt.subplot(1, 3, 2)
sns.histplot(df['Current Price ($)'], kde=True, bins=10, edgecolor='k', color="green")
plt.xlabel('Current Price')
plt.ylabel('Frequency')
plt.title('Histogram of Current Price')

plt.subplot(1, 3, 3)
sns.histplot(df['Gain/Loss ($)'], kde=True, bins=10, edgecolor='k', color="brown")
plt.xlabel('Gain/Loss')
plt.ylabel('Frequency')
plt.title('Histogram of Gain/Loss')

plt.tight_layout()
plt.show()
```

```python
# Sort data by Dividend / Shares
data = df.sort_values(by="Dividend/Shares", ascending=False)
# Get the 5th highest Dividend / Shares
top_5 = data.head(5)
# Plot the Data
plt.figure(figsize=(9, 4))
sns.barplot(x="Stock Name", y="Dividend/Shares", data=top_5)
plt.title("Top 5 Stocks with Highest Dividend / Shares")
plt.show()
```

```python
# Sort data by Market Value ($)
data = df.sort_values(by="Market Value ($)", ascending=False)
# Get the 5th highest Market Value ($)
top_5 = data.head(5)

# Generate a color palette with the number of lines
num_lines = len(top_5)
color_palette = sns.color_palette("Set1", num_lines)

# Plot the data
plt.figure(figsize=(9, 4))
for i in range(num_lines):
    plt.plot(top_5["Stock Name"].iloc[i], top_5["Market Value ($)"].iloc[i], marker='o', color=color_palette[i])

plt.title("Top 5 Stocks with Highest Market Value ($)")
plt.xlabel("Stock Name")
plt.ylabel("Market Value ($)")
plt.xticks(rotation=45)
plt.show()
```

```python
# Sort data by Annual Yield (%)
data = df.sort_values(by="Annual Yield (%)", ascending=False)
# Get the 5th highest Annual Yield (%)
top_7 = data.head(7)

# Generate a color palette with the number of stocks
num_stocks = len(top_7)
color_palette = sns.color_palette("Set1", num_stocks)

# Plot the data
plt.figure(figsize=(13, 4))
plt.scatter(top_7["Stock Name"], top_7["Annual Yield (%)"], c=color_palette)
plt.title("Top 7 Stocks with Highest Annual Yield (%)")
plt.xlabel("Stock Name")
plt.ylabel("Annual Yield (%)")
plt.xticks(rotation=45)
plt.show()
```
